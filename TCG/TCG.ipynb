```python
tipocarta = dict(collections.Counter(TC['Race']))
total= dict()
total['Races']=list(tipocarta.keys())
total['Count']=list(tipocarta.values())
```

```python
total = pd.DataFrame(total)
print(total)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
plt.bar(total['Races'], total['Count'])
plt.title('Tipos de cartas')
plt.xlabel('Tipo')
plt.ylabel('Total')

plt.xticks(rotation=90)
plt.show()
```
